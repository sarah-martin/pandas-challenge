### Data Analysis
* There seems to be a negative correlation to school size and overall passing grade, meaning students are less likely to get higher grades with an increased number of classmates.
* Charter school students have the highest chance of obtaining higher grades, likely due to the smaller class sizes.
* There is not a positive correlation between budget per student and overall passing percentage. 

```python
# Dependencies and Setup
import pandas as pd

# Load files
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)

# Combine the data into a single dataset
data_complete = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])
```

   ### District Summary

```python
# Calculate total number of school by pulling the unique instances and putting in list
schools_total=data_complete["school_name"].unique()

# Calculate total number of students by counting number in file
students_total=data_complete["student_name"].count()

# Calculate total budget by summing unique instances
budget_list=data_complete["budget"].unique()
budget_total=sum(budget_list)

# Calculate math and reading score means
math_avg=data_complete["math_score"].mean()
reading_avg=data_complete["reading_score"].mean()

# Calculate average math and reading scores
pass_math_total=(data_complete["math_score"]>=70).sum()
pass_math_percent=(pass_math_total/students_total)*100

pass_read_total=(data_complete["reading_score"]>=70).sum()
pass_read_percent=(pass_read_total/students_total)*100

pass_overall=data_complete[(data_complete["math_score"]>=70) & (data_complete["reading_score"]>=70)].count()
pass_percent=(pass_overall/students_total)*100
                                
#Create a summary dataframe
summary_df=pd.DataFrame({"Total Schools": len(schools_total),
                         "Total Students":[students_total],
                         "Total Budget": [budget_total],
                         "Average Math Score": [math_avg],
                         "Average Reading Score": [reading_avg],
                         "% Passing Math": [pass_math_percent],
                         "% Passing Reading": [pass_read_percent],
                         "% Overall Passing": [pass_percent]})
summary_df
```

### School Summary

```python
# Group by school
school_group = data_complete.groupby(["school_name"])

# Establish type of school by using only the first instance of the record
school_type=school_group["type"].first()

# Count total number of students
grouped_student_total=school_group["student_name"].count()

# Find total budget per school
grouped_total_budget=school_group["budget"].first()

# Calculate budget per student per school
grouped_student_budget=grouped_total_budget/grouped_student_total

# Calculate average math and reading scores using average function
grouped_math_avg=school_group["math_score"].mean()
grouped_read_avg=school_group["reading_score"].mean()

# Find and calcuate to find total passing each subject
grouped_math_pass_total=data_complete[data_complete["math_score"]>=70].groupby("school_name")["Student ID"].count()
grouped_math_pass_pct=(grouped_math_pass_total/grouped_student_total)*100

grouped_read_pass_total=data_complete[data_complete["reading_score"]>=70].groupby("school_name")["Student ID"].count()
grouped_read_pass_pct=(grouped_read_pass_total/grouped_student_total)*100

# Find and calculate total overall passing 
grouped_pass=data_complete[(data_complete["math_score"]>=70) & (data_complete["reading_score"]>=70)].groupby("school_name")["Student ID"].count()
grouped_pass_total=(grouped_pass/grouped_student_total)*100

# Create dataframe for summary information
school_summary_df=pd.DataFrame({"School Type": school_type,
                                "Total Students": grouped_student_total,
                                "Total School Budget": grouped_total_budget,
                                "Per Student Budget": grouped_student_budget,
                                "Average Math Score": grouped_math_avg,
                                "Average Reading Score": grouped_read_avg,
                                "% Passing Math": grouped_math_pass_pct,
                                "% Passing Reading": grouped_read_pass_pct,
                                "% Overall Passing": grouped_pass_total})
school_summary_df
```

### Top Performing Schools (By % Overall Passing)

```python
# Sort and display the top five performing schools by % overall passing.
top_performing_df=school_summary_df.sort_values(("% Overall Passing"), ascending=False)
top_performing_df.head(5)
```

### Bottom Performing Schools (By % Overall Passing)

```python
# Sort and display the five worst-performing schools by % overall passing.
lowest_performing_df=school_summary_df.sort_values(("% Overall Passing"))
lowest_performing_df.head(5)
```

### Math Scores by Grade

```python
#Calculate average math scores per grade, grouped by school
math_avg_9=data_complete[data_complete["grade"]=="9th"].groupby("school_name")["math_score"].mean()
math_avg_10=data_complete[data_complete["grade"]=="10th"].groupby("school_name")["math_score"].mean()
math_avg_11=data_complete[data_complete["grade"]=="11th"].groupby("school_name")["math_score"].mean()
math_avg_12=data_complete[data_complete["grade"]=="12th"].groupby("school_name")["math_score"].mean()

#Create the dataframe
grade_math_df=pd.DataFrame({"9th": math_avg_9,
                            "10th": math_avg_10,
                            "11th": math_avg_11,
                            "12th": math_avg_12})
grade_math_df
```

### Reading Score by Grade

```python
#Calculate average reading scores per grade, grouped by school
read_avg_9=data_complete[data_complete["grade"]=="9th"].groupby("school_name")["reading_score"].mean()
read_avg_10=data_complete[data_complete["grade"]=="10th"].groupby("school_name")["reading_score"].mean()
read_avg_11=data_complete[data_complete["grade"]=="11th"].groupby("school_name")["reading_score"].mean()
read_avg_12=data_complete[data_complete["grade"]=="12th"].groupby("school_name")["reading_score"].mean()

#Create the dataframe
grade_read_df=pd.DataFrame({"9th": read_avg_9,
                            "10th": read_avg_10,
                            "11th": read_avg_11,
                            "12th": read_avg_12})
grade_read_df
```

### Scores by School Spending

```python
#Create and label bins for per student budget
budget_bin=[0,585,610,635,660]
budget_labels=["< $585","$585-609","$610-634","$635-659"]

#Associate budget with bins
school_summary_df[("Spending Ranges (Per Student)")]=pd.cut(school_summary_df["Per Student Budget"],budget_bin,labels=budget_labels)

#Group the datagrame by the bins
budget_df=school_summary_df.groupby("Spending Ranges (Per Student)")

#Form dataframe and find the averages
budget_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean()
```

### Scores by School Size

```python
#Create and label bins for per student budget
size_bin=[0,1000,2000,5000]
size_labels=["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"]

#Associate budget with bins
school_summary_df[("School Size")]=pd.cut(school_summary_df["Total Students"],size_bin,labels=size_labels)

#Group the datagrame by the bins
size_df=school_summary_df.groupby("School Size")

#Form dataframe and find the averages
size_df[["Average Math Score","Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]].mean()
```
